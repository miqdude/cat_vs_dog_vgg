# Dataframe Class

```python
import torch.nn as nn
from torch.utils.data import DataLoader
from torch.utils.data import Dataset
import torch.nn.functional as F
import torch
import numpy as np
from tqdm.notebook import tqdm
from sklearn.model_selection import train_test_split
```

```python
class CatDog(Dataset):
    def __init__(self, source_folder, file_formats):
        self.source_folder = source_folder
        self.file_formats = file_formats
        
        self.filenames = [file for file in os.listdir(source_folder) if file.endswith(file_formats)]
        self.train, self.test = train_test_split(self.filenames)
        
        self.target_split = self.train
                 
    def __len__(self):
        return len(self.target_split)
                 
    def set_mode(self, train):
        if train == "train":
            self.target_split = self.train
        elif train == "test":
            self.target_split = self.test
            
    def __getitem__(self, index):
        label = self.target_split[index][0:3]
        if label == "cat" :
            label = 0
        elif label == "dog" :
            label = 1

        img  = cv2.imread(self.source_folder + "/" + self.target_split[index]).astype("float32").reshape((3,224,224))
        row = torch.from_numpy(img)
        
        return {
            "x_data" : row,
            "y_label" : label
        }
    
    def get_num_batches(self, num_batches):
        return len(self.target_split)//num_batches
```

```python
def generate_batches(dataset, batch_size, device):
    dataloader = DataLoader(dataset, batch_size=batch_size)
    
    for data_dict in dataloader:
        out_data_dict = {}
        for name, tensor in data_dict.items():
            out_data_dict[name] = data_dict[name]
        yield out_data_dict
```

# Network Class

```python
class CATDOGClassifier(nn.Module):
    def __init__(self):
        super(CATDOGClassifier, self).__init__()
        
        self.conv2d1 = nn.Conv2d(3, 64, 3, 1, 1)
        self.conv2d2 = nn.Conv2d(64, 64, 3, 1, 1)
        self.conv2d3 = nn.Conv2d(64, 128, 3, 1, 1)
        self.conv2d4 = nn.Conv2d(128, 128, 3, 1, 1)
        self.conv2d5 = nn.Conv2d(128, 256, 3, 1, 1)
        self.conv2d6 = nn.Conv2d(256, 256, 3, 1, 1)
        self.conv2d7 = nn.Conv2d(256, 512, 3, 1, 1)
        self.conv2d8 = nn.Conv2d(512, 512, 3, 1, 1)
        
        
        self.fcs = nn.Sequential(
            nn.Linear(512 * 7 * 7, 4096),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            nn.Linear(4096, 4096),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            nn.Linear(4096, 2),
        )
        
        self.maxpool = nn.MaxPool2d(kernel_size=(2,2),stride=(2,2))
        
    def forward(self, x):
        x = self.conv2d1(x) # 224 224 64
        x = F.relu(x)
        x = self.conv2d2(x) # 224 224 64
        x = F.relu(x)
        
        x = self.maxpool(x)
        x = self.conv2d3(x) # 112 112 128
        x = F.relu(x)
        x = self.conv2d4(x) # 112 112 128
        x = F.relu(x)
        
        # 56 x 56 x 256
        x = self.maxpool(x)
        x = self.conv2d5(x) # 56 56 256
        x = F.relu(x)
        x = self.conv2d6(x) # 56 56 256
        x = F.relu(x)
        x = self.conv2d6(x) # 56 56 256
        x = F.relu(x)
        x = self.conv2d6(x) # 56 56 256
        x = F.relu(x)
        
        # 28 x 28 x 512
        x = self.maxpool(x)
        x = self.conv2d7(x) # 28 28 512 
        x = F.relu(x)
        x = self.conv2d8(x) # 28 28 512
        x = F.relu(x)
        x = self.conv2d8(x) # 28 28 512
        x = F.relu(x)
        
        x = self.maxpool(x)
        x = self.conv2d8(x) # 14 14 512
        x = F.relu(x)
        x = self.conv2d8(x) # 14 14 512
        x = F.relu(x)
        x = self.conv2d8(x) # 14 14 512
        x = F.relu(x)
        
        x = self.maxpool(x)
        # after maxpool needs to reshape
        # since the paper did not mention it
        x = x.reshape(x.shape[0], -1)
        x = self.fcs(x)
        
        output = F.softmax(x, 1)
        return output
        
```

# Training

```python
def sequence_loss(y_pred, y_true):
    return F.cross_entropy(y_pred, y_true)


def compute_accuracy(y_pred, y_true):

    _, y_pred_indices = y_pred.max(dim=1)
    
    # print(y_pred_indices.shape)

    correct_indices = torch.eq(y_pred_indices, y_true).int().sum()

    return correct_indices / len(y_true) * 100
```

```python
import torch.optim as optim

device = "cuda" if torch.cuda.is_available() else "cpu"

model = CATDOGClassifier()
model.to(device)

# optimizer = optim.Adam(model.parameters(), lr=.001)
optimizer = optim.SGD(model.parameters(), lr=.001, momentum=.9)

source_folder = "/content/drive/MyDrive/Colab Notebooks/CatVsDog/train_resized"
file_formats = ("jpg")
dataset = CatDog(source_folder, file_formats)
```

```python
"""
    In Collab Notebook the available GPU memory is 11GB
    this network uses reserved 10GB so there is no much room left
"""

epoch = 74
batch_size = 64

train_bar = tqdm(desc='iteration routine', 
                          total=dataset.get_num_batches(batch_size),
                          position=0)
epoch_bar = tqdm(desc='training routine', 
                          total=epoch,
                          position=0)
try:
    loss_list = []
    accuracy_list = []
    for epoch in range(epoch):
        batch_generator = generate_batches(dataset, batch_size=batch_size, device=device)
        model.train()

        for batch_index, batch_dict in enumerate(batch_generator):
            # print("Batch %d-th" % batch_index)
            optimizer.zero_grad()

            y_pred = model(batch_dict["x_data"])

            loss = sequence_loss(y_pred, batch_dict["y_label"])

            loss.backward()

            optimizer.step()

            accuracy = compute_accuracy(y_pred, batch_dict["y_label"])

            loss_list.append(loss)
            accuracy_list.append(accuracy)
            
            # print("Train %d loss %.2f" % (batch_index, loss))
            # print("Train %d accuracy %.2f" % (batch_index, accuracy))

            train_bar.set_postfix(
                loss = loss,
                accuracy = accuracy,
                epoch = batch_index,
            )
            train_bar.update()

        epoch_bar.update()

except KeyboardInterrupt:
    print("Keyboard interrupt")
```
